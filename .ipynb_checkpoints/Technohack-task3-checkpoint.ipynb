#  Visualization using Histogram

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data=pd.read_csv('Iris.csv')
```

```python
data.head()
```

```python
data.drop('Id',axis=1, inplace=True) 
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#check for duplicated values
len(data[data.duplicated()])
```

```python
data.Species.value_counts()
```

```python
#scatter plot X axix for sepalLength and y axis for sepalWidth 
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm', data=data,hue='Species')
```

```python
sns.scatterplot(x='PetalLengthCm',y='PetalWidthCm', data=data,hue='Species')
```

```python
for col in data.columns[:-1]:
    sns.boxplot(x='Species',y=col,data=data)
    plt.show()
```

```python
for col in data.columns[:-1]:
    sns.histplot(x=col,data=data)
    plt.show()
```

```python
sns.pairplot(data=data,hue='Species')
```
